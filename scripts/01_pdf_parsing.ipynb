```python
import fitz

pdf_path = "../data/fta-cn/psr_kor.pdf"  # 경로 수정

doc = fitz.open(pdf_path)

print("페이지 수:", len(doc))
```

```python
text = doc[0].get_text()

print(text[:3000])
```

```python
all_text = ""

for page in doc:
    all_text += page.get_text()

print("총 글자 수:", len(all_text))
```

```python
with open("../output/psr_kor.txt", "w", encoding="utf-8") as f:
    f.write(all_text)

print("저장 완료")
```

```python
import re

codes = re.findall(r"\d{4}\.\d{2}", all_text)

print("추출 개수:", len(codes))
print("처음 20개:")
print(codes[:20])
```
