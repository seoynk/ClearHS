```python
#PDF 로드

import fitz
import re
import pandas as pd

pdf_path = "../data/fta-cn/psr_kor.pdf"

doc = fitz.open(pdf_path)

all_text = ""

for page in doc:
    all_text += page.get_text()

print("총 글자수:", len(all_text))
```

```python
#HS 코드 위치 찾기

pattern = r"\d{4}\.\d{2}"

matches = list(re.finditer(pattern, all_text))

print("HS 코드 개수:", len(matches))

for m in matches[:10]:
    print(m.group())
```

```python
#chunk 생성

chunks = []

for i in range(len(matches)):

    start = matches[i].start()

    if i < len(matches) - 1:
        end = matches[i + 1].start()
    else:
        end = len(all_text)

    hs_code = matches[i].group()

    content = all_text[start:end].strip()

    chunks.append({
        "hs_code": hs_code,
        "content": content
    })

print("생성된 chunk 수:", len(chunks))
```

```python
import re

def is_psr_code(line):
    line = line.strip()

    patterns = [
    r'^\d{4}$',
    r'^\d{4}\.\d{2}$',
    r'^\d{4}\s*[-–]\s*\d{4}$',
    r'^\d{4}\.\d{2}\s*[-–]\s*\d{4}\.\d{2}$',
    r'^\d{4}\s*[-–]\s*\d{4}\.\d{2}$',
]

    return any(re.match(p, line) for p in patterns)
```

```python
lines = all_text.split("\n")

for line in lines:
    if is_psr_code(line):
        print(line)
```

```python
codes = []

for i, line in enumerate(lines):
    if is_psr_code(line):
        codes.append((i, line.strip()))

print("코드 수:", len(codes))
codes[:20]
```

```python
chunks = []

for idx in range(len(codes)):

    start_line = codes[idx][0]
    code = codes[idx][1]

    if idx < len(codes) - 1:
        end_line = codes[idx + 1][0]
    else:
        end_line = len(lines)

    content = "\n".join(lines[start_line:end_line])

    chunks.append({
        "code": code,
        "content": content
    })

print("chunk 수:", len(chunks))
```
