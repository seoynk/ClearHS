```python
import fitz

pdf_path = "../data/fta-us/21-07CustomsAdmin_kor.pdf"

doc = fitz.open(pdf_path)

all_text = ""

for page in doc:
    all_text += page.get_text()

print(all_text[:5000])
```

```python
lines = all_text.split("\n")

for i in range(200):
    print(repr(lines[i]))
```

```python
import re

articles = []

for i, line in enumerate(lines):

    line = line.strip()

    if re.match(r'^제\d+\.\d+\s*조$', line):
        articles.append((i, line))

print("조문 수:", len(articles))
articles[:10]
```

```python
# 청킹

chunks = []

for idx in range(len(articles)):

    start = articles[idx][0]

    article = articles[idx][1]

    # 제목
    title = lines[start + 1].strip()

    # 다음 조문 위치
    if idx < len(articles) - 1:
        end = articles[idx + 1][0]
    else:
        end = len(lines)

    content = "\n".join(lines[start:end])

    chunks.append({
        "agreement": "KR_CN_FTA",
        "chapter": "통관절차 및 무역원활화",
        "article": article,
        "title": title,
        "content": content
    })

print("청크 수:", len(chunks))
print(chunks[0])
```

```python
import pandas as pd

df = pd.DataFrame(chunks)

df.head()
```

```python
df.to_csv(
    "../output/kr_us_customs_facilitation.csv",
    index=False,
    encoding="utf-8-sig"
)

print("저장 완료")
```
